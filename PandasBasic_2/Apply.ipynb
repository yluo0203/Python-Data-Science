```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
df = pd.read_csv("apply_demo.csv")
df.shape
```

```python
df.head()
```

```python
df["A"] = Series(["a"] * 3989)
df.head()
```

```python
df["A"] = df["A"].apply(str.upper)
df.head()
```

```python
del df["A"]
df.head()
```

```python
temp = df["data"][0].split("")
temp
```

```python
temp[-1]
```

```python
def a_new_line(line):
    line = line.strip().split(" ")
    return Series([line[1], line[3], line[5]])
```

```python
df_temp = df["data"].apply(a_new_line)
df_temp.head()
```

```python
df_temp = df_temp.rename(columns = {0: "Name", 1:"Seq", 2:"Price"})
df_temp.head()
```

```python
df_new = df.combine_first(df_temp)
df_new.head()
```

```python
del df_new["data"]
df_new
```
