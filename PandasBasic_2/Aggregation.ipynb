# Aggregation

```python
g = df.groupby(["city"])
g.groups
```

```python
g.agg("max")
# a function "max"
```

```python
# Define a function
def Wind_diff(attr):
    return attr.max() - attr.min()

g.agg(Wind_diff)
```
