```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
s = Series(np.random.randn(10000))
s.plot(kind = "hist")
# plt.hist(s)
```

```python
s.plot(kind = "kde")
```

```python
sns.distplot(s)
```

```python
sns.distplot(s, kde = False, color = "r", bins = 50)
```

```python
sns.kdeplot(s, color = "r")
```

```python
sns.kdeplot(s, color = "r", shade = True)
```
